```python
import requests
import time
import csv
from bs4 import BeautifulSoup
```

```python
articles_data = []
company_name = '라인'
```

```python
for page_num in range(1,39):
    if page_num == 1:
        list_page_res = requests.get('https://engineering.linecorp.com/ko/blog')
    else:
        list_page_res = requests.get('https://engineering.linecorp.com/ko/blog/page/{}'.format(page_num))
    list_page_res.encoding = 'UTF-8'
    list_page_soup = BeautifulSoup(list_page_res.text, 'html.parser')
    posts = list_page_soup.find('ul', class_='post_list').find_all('li', class_='post_list_item')

    for post in posts:
        tags_list = []
        tags = post.find_all('li')

        for tag in tags:
            tags_list.append(tag.text)

        article_data = {
            'title' : post.find('h2').text,
            'date' : post.find('span').text,
            'tags' : tags_list,
            'company' : company_name,
            'link' : 'https://engineering.linecorp.com' + post.find('a')['href']
        }
        articles_data.append(article_data)

    time.sleep(5)

print(len(articles_data))
```

```python
file_name = './full_line_data.csv'
with open(file_name, mode='w', newline='', encoding='utf-8') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=articles_data[0].keys())
    
    writer.writeheader()
    
    for data in articles_data:
        data['tags'] = ', '.join(data['tags'])
        writer.writerow(data)

print('완료')
```
