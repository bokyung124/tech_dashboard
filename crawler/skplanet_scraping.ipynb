```python
import requests
import time
import csv
from bs4 import BeautifulSoup
```

```python
articles_data = []
company_name = 'SK플래닛'
```

```python
res = requests.get("https://techtopic.skplanet.com/")
soup = BeautifulSoup(res.text, "html.parser")
```

```python
posts = soup.find('ol', class_='list').find_all('li')
for post in posts:
    tag_list = []
    tags = post.find('div', class_='tags')
    for tag in tags:
        tag_list = []
        tag_list.append(tag.text)

    article_data = {
        'title' : post.find('h2').text,
        'date' : post.find('small').text[:10],
        'tags' : tag_list,
        'company' : company_name,
        'link' : 'https://techtopic.skplanet.com/' + post.find('h2').find('a')['href']
    }
    articles_data.append(article_data)

print(len(articles_data))
```

```python
file_name = './skplanet_data.csv'
with open(file_name, mode='w', newline='', encoding='utf-8') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=articles_data[0].keys())
    
    writer.writeheader()
    
    for data in articles_data:
        data['tags'] = ', '.join(data['tags'])
        writer.writerow(data)

print('완료')
```
