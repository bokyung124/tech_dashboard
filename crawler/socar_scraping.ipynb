```python
import requests
import time
import csv
from bs4 import BeautifulSoup
```

```python
articles_data = []
company_name = '쏘카'
```

본코드

```python
for page_num in range(1,11):
# for page_num in range(1,2):
    if page_num == 1:
        list_page_res = requests.get('https://tech.socarcorp.kr/posts/')
    else:
        list_page_res = requests.get('https://tech.socarcorp.kr/posts/page{}/'.format(page_num))
    # list_page_res.encoding = 'UTF-8'
    list_page_soup = BeautifulSoup(list_page_res.text, 'html.parser')
    posts = list_page_soup.find_all('article', class_='post-preview')

    for post in posts:
        tags_list = []
        tags = post.find_all('span', class_='tag')

        for tag in tags:
            tags_list.append(tag.text)
        print(tags_list)

        article_data = {
            'title' : post.find('h2').text,
            'date' : post.find('span', class_='date').text,
            'tags' : tags_list,
            'company' : company_name,
            'link' : 'https://tech.socarcorp.kr' + post.find('a')['href']
        }
        articles_data.append(article_data)

    time.sleep(10)

print(len(articles_data))
```

```python
file_name = './socar_data.csv'
with open(file_name, mode='w', newline='', encoding='utf-8') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=articles_data[0].keys())
    
    writer.writeheader()
    
    for data in articles_data:
        data['tags'] = ', '.join(data['tags'])
        writer.writerow(data)

print('완료')
```
